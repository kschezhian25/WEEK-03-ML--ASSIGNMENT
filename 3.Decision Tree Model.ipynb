```python
#Import Libraries
import pandas as pd
#Read the excel file
dataset=pd.read_csv("insurance_pre.csv")
#data set is in categorical data- conver to numerical using dummies
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
#data set is in categorical data- conver to numerical using dummies
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
#Intependent variable Or input variable
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
independent
```

```python
#dependent variable  OR Output Variable
dependent=dataset[["charges"]]
```

```python
#Train and Test(Split)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation.created using training set -------------->TC-1<----------------------------------------
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor(criterion='squared_error',splitter='best',max_features=None)
regressor.fit(x_train, y_train)
############regressor will hold the model---fit method
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
#predicting
y_pred=regressor.predict(x_test)
#validating
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
